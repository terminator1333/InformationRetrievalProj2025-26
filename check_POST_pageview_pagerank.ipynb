Check Pageview

```python
import requests

#our GCP external IP and the port defined in search_frontend.py (8080)
url = 'http://35.226.215.146:8080/get_pageview'

# List of wiki IDs to query (e.g., Barack Obama is often ID 534366, or generic IDs)
# Make sure to use IDs that actually exist in your pageviews.pkl for a non-zero result.
payload = [i for i in range(20)]

try:
    # sending POST request
    response = requests.post(url, json=payload)

    if response.status_code == 200:
        print("Success!")
        print("Pageviews:", response.json())
    else:
        print(f"Failed with status code: {response.status_code}")
        print("Response:", response.text)

except requests.exceptions.ConnectionError:
    print("Could not connect")
```

Check Pagerank

```python
import requests

# Our GCP external IP and port 8080
url = 'http://35.226.215.146:8080/get_pagerank'

# List of wiki IDs to query
payload = [i + 1000 for i in range(20)]

try:
    # Sending the POST request
    response = requests.post(url, json=payload)

    if response.status_code == 200:
        print("Success!")
        # The backend returns a list of floats corresponding to the IDs
        print("PageRanks:", response.json())
    else:
        print(f"Failed with status code: {response.status_code}")
        print("Response:", response.text)

except requests.exceptions.ConnectionError:
    print("Could not connect. Check if the server is running and the Firewall rules allow port 8080.")
```
